# Statistics by Group

```python
#Extract female valued rows from the dataframe
female = df[df["gender"] == "female"]
#Describe the dataframe
describe_f = female.describe()
#Extract eval, beauty and age
describe_f = describe_f[["eval", "beauty", "age"]]
print(describe_f)
```

```python
#The Lazy Version
female = df[df["gender"] == "female"].describe()[["eval", "beauty", "age"]]


male = df[df["gender"] == "male"].describe()[["eval", "beauty", "age"]]
minority = df[df["minority"] == "yes"].describe()[["eval", "beauty", "age"]]
n_minority = df[df["minority"] == "no"].describe()[["eval", "beauty", "age"]]
tenured_prof = df[df["tenured_prof"] == 1].describe()[["eval", "beauty", "age"]]
n_tenured_prof = df[df["tenured_prof"] == 0].describe()[["eval", "beauty", "age"]]
```

```python
print( "\nMale Instructor\n", male, "\nFemale Instructor\n",female,
      "\nNon-Tenured Professor\n",n_tenured_prof, "\nTenured Professor\n", tenured_prof,
     "\nVisible Minority Instructor\n", minority, "\nVisible Minority Instructor\n", n_minority)
```

```python
#Extract female valued rows from the dataframe
female = df[df["gender"] == "female"]
print("Female professor only dataframe\n\n", female.head(),"\n\n ")# \n is for easy to view formatting

#Find duplicated values with selected columns. The output is series of boolean values- True means it is a duplicate
dups = fem.duplicated(subset = ["age", "beauty"])
print("duplications list\n\n", dups.head(),"\n\n") #True means it is a duplicate

#Reverse the boolean series to use for indexing
reverse_dups = -dups
print("reversed duplications list\n\n", reverse_dups.head(),"\n\n")

#Extract not duplicate values from dataframe
female_non_dup = female[reverse_dups]
print("Female professors dataframe without the duplicates\n\n", female_non_dup.head(),"\n\n")

#Describe the dataframe from selected columns
female_non_dup_describe = female_non_dup.describe()[["eval", "age", "beauty"]]
print("Female professors dataframe without the duplicates description\n\n", female_non_dup_describe)

```

```python
#The lazy version
fem = df[df["gender"] == "female"] 
fem_des = fem[-(fem.duplicated(subset = ["age", "beauty"]))].describe()[["eval", "age", "beauty"]]
print("Female professors dataframe without the duplicates description\n\n",fem_des)
```
